```python
import zipfile
import json
import re

# Brand extraction function: extract @xxx from caption
def extract_brand_from_jsonfile(file):
    try:
        data = json.load(file)
        edges = data.get("edge_media_to_caption", {}).get("edges", [])
        if edges:
            caption = edges[0]["node"]["text"]
            matches = re.findall(r"@(\w+)", caption)
            if matches:
                return matches[0]
        return "no_brand"
    except:
        return "read_error"
```

```python
zip_path = "/content/drive/MyDrive/Colab Notebooks/json_files.zip"

brands = []
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
   # Only process matching json files
    for json_name in filtered_df[3]:
        try:
            with zip_ref.open(f"json/{json_name}") as file:
                brand = extract_brand_from_jsonfile(file)
                brands.append(brand)
        except:
            brands.append("not_found")
```

```python
def smart_extract_brand(file):
    try:
        data = json.load(file)

        # 1. sponsor_user
        sponsor_edges = data.get("edge_media_to_sponsor_user", {}).get("edges", [])
        if sponsor_edges:
            sponsor = sponsor_edges[0]["node"].get("sponsor", {}).get("username")
            if sponsor:
                return sponsor

        # 2. caption 中找 @xxx
        caption_edges = data.get("edge_media_to_caption", {}).get("edges", [])
        if caption_edges:
            caption = caption_edges[0]["node"]["text"]
            matches = re.findall(r"@(\w+)", caption)
            if matches:
                return matches[0]

        return "no_brand"
    except:
        return "read_error"
# Extract again using the smart_extract_brand function
updated_brands = []

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    for json_name in unmatched_df[3]:
        try:
            with zip_ref.open(f"json/{json_name}") as file:
                brand = smart_extract_brand(file)
                updated_brands.append(brand)
        except:
            updated_brands.append("not_found")
unmatched_df["brand"] = updated_brands

# Save the updated results
unmatched_df.to_csv("no_brand_updated_smart.csv", index=False)
```

```python
import zipfile
import pandas as pd
import os


zip_path_brand = "/content/drive/MyDrive/Colab Notebooks/profiles_brands.zip"

brand_data = []
with zipfile.ZipFile(zip_path_brand, 'r') as zip_ref:
    brand_files = [f for f in zip_ref.namelist() if f.startswith("users_brands_SPOD/") and len(f.strip().split("/")) == 2]

    for file in brand_files:
        try:
            with zip_ref.open(file) as f:
                content = f.read().decode("utf-8", errors="ignore").strip()
                fields = content.split('\t')
                fields += [""] * (11 - len(fields))

                brand_data.append({
                    "file_name": os.path.basename(file),
                    "Name": fields[0],
                    "Followers": fields[1],
                    "Followees": fields[2],
                    "Posts": fields[3],
                    "URL": fields[4],
                    "T/F": fields[5],
                    "Category": fields[6],
                    "Bio": fields[7],
                    "Email": fields[8],
                    "Phone": fields[9],
                    "Profile_pic": fields[10]
                })
        except Exception as e:
            print(f"❌ Failed to parse {file}: {e}")

# 构建 DataFrame 并导出
brand_profiles_df = pd.DataFrame(brand_data)

brand_profiles_df.to_csv("brand_profiles.csv", index=False)


```

```python
from sklearn.model_selection import train_test_split
import pandas as pd



# Get the unique internet celebrity account
owners = df['owner'].unique()

# Divide the owner list of train / val / test
train_owners, temp_owners = train_test_split(owners, test_size=0.3, random_state=42)
val_owners, test_owners = train_test_split(temp_owners, test_size=0.5, random_state=42)

# Label each influencer
def assign_split(owner):
    if owner in train_owners:
        return 'train'
    elif owner in val_owners:
        return 'val'
    else:
        return 'test'

df['split'] = df['owner'].apply(assign_split)

# Save or use the divided data
df.to_csv("influencer_soft_labels_with_split.csv", index=False)
```
